# Matplotlib

## What is Matplotlib? {.hide .smaller-85}

**Matplotlib** is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for scientific computing and data visualization. Key features include:

- High-quality 2D and 3D plots
- Support for various output formats (e.g., PDF, PNG, SVG)
- Fine-grained control over all aspects of presentation

## Rendering Environment {.hide}

Matplotlib can render plots in various environments. Sometimes, it may assume a GUI is available when it is not. To avoid issues on systems without a display, set the backend to `Agg`. This non-interactive backend is ideal for generating plots as image files or for batch processing. 

```python
import matplotlib
matplotlib.use('Agg')  # Use 'Agg' for non-GUI environments
```

::: {.callout-note}
**Agg** stands for Anti-Grain Geometry, it is a high-quality rendering engine for raster graphics.
:::

## Plotting with Matplotlib {.hide}

Matplotlib provides two interfaces for creating plots:

- **MATLAB-style interface**: Convenient for quick and simple plots.
- **Object-oriented interface**: Offers more control and flexibility over plot elements.


## MATLAB-style Interface {.hide}

The MATLAB-style interface is provided by the `pyplot` module. It is a collection of functions that make Matplotlib work like MATLAB.

The standard way to import the `pyplot` module is as follows:

```python
import matplotlib.pyplot as plt
```

## Example: MATLAB-style Interface {.hide .smaller-70}

This demonstrates the MATLAB-style interface.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y, 'b-', linewidth=2)
plt.show()
````

## Example: Object-oriented Interface {.hide .smaller-70}

The object-oriented interface is more powerful and flexible than the MATLAB-style interface. It involves creating figure and axis objects explicitly and calling methods on them.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'b-', linewidth=2)
plt.show()
````

## Object-oriented Interface Explained {.hide .smaller-70}


::: {.columns}

::: {.column width=50%}
`fig, ax = plt.subplots()` returns:

- Figure (`fig`), which is the blank canvas.
- AxesSubplot (`ax`), which is a frame for plotting in.

Notice the `plot()` function is a method of `ax`.
:::

::: {.column width=50%}
```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```
:::
:::


## Legend {.hide}

:::{.smaller-70}
The `legend()` method adds a legend to the plot. It takes an optional `loc` argument to specify the location of the legend. The `loc` argument can be a string (e.g., 'upper left'), a tuple of coordinates (e.g., (0.5, 0.5)), or a number (e.g., 2).
:::

````python
fig, ax = plt.subplots()
ax.plot(x, y, 'r-', linewidth=2, label='sine function', alpha=0.6)
ax.legend(loc=(0.05, 0.1))
plt.show()
````

## Ticks, Tick Labels, and Titles (.hide)

You can customize the ticks and titles of the plot using the `set_xticks()`, `set_xticklabels()`, `set_xlabel()`, and `set_title()` methods.

````python
fig, ax = plt.subplots()
ax.plot(x, y, 'r-', linewidth=2, label='sine function', alpha=0.6)
ax.legend(loc='upper center')
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels(['min', 'mid', 'max'])
ax.set_title('Test plot')
plt.show()
````
